#### Synthetic time series

```r
x <- seq(0, 10, 0.25)
data <- data.frame(x, sin=sin(x))
head(data)
```

## ts plot

A time series plot during exploratory analysis

```r
grf <- plot_ts(x = data$x, y = data$sin, color=colors[1])
plot(grf)
```
